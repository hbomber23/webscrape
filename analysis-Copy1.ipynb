```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', 500)

temp=pd.read_csv('SI_therapists.csv', encoding='latin-1')
df_SI = pd.DataFrame(temp)
df_SI

temp=pd.read_csv('QNS_therapists.csv', encoding='latin-1')
df_QNS = pd.DataFrame(temp)
df_QNS

temp=pd.read_csv('BX_therapists.csv', encoding='latin-1')
df_BX = pd.DataFrame(temp)
df_BX

temp=pd.read_csv('BK_therapists.csv', encoding='latin-1')
df_BK = pd.DataFrame(temp)
df_BK

temp=pd.read_csv('MN_therapists.csv', encoding='latin-1')
df_MN = pd.DataFrame(temp)
df_MN
```
